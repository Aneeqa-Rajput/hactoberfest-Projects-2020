```python
def space_check(board, position):
    

    return board[position] == ' '
```

```python
def full_board_check(board):
    
    return ' ' not in board[1:10]
```

```python
def player_choice(board):
    
    
    pos = int(input("Enter the position where you want to insert : "))
    
    if space_check(board,pos):
        return pos
    else:
        print("Sorry , Position is filled")
        player_choice(board)
```

```python
def replay():
    
    reply = input("Want to play again? (yes/no)").lower()
    
    return reply == 'yes'
```
